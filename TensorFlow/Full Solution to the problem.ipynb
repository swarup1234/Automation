```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Reading the relavant data frame
npz = np.load("Audiobooks_data_train.npz")
train_inputs = npz['inputs'].astype(np.float)
train_targets = npz['targets'].astype(np.int)

# Reading validation and test data
npz = np.load("Audiobooks_data_validation.npz")
validation_inputs, validation_targets = npz['inputs'].astype(np.float), npz['targets'].astype(np.int)
npz = np.load("Audiobooks_data_test.npz")
test_inputs, test_targets = npz['inputs'].astype(np.float), npz['targets'].astype(np.int)
```

```python
test_inputs
```

```python
# Model of the data as discussed before
input_size = 10
hidden_layer_size = 50
output_size = 2

model = tf.keras.Sequential([
                            tf.keras.layers.Dense(hidden_layer_size, activation = 'relu'),
                            tf.keras.layers.Dense(hidden_layer_size, activation = 'relu'),
                            tf.keras.layers.Dense(output_size, activation = 'softmax')
                            ])

# one hot encoded during calculating the entropy - sparse cat entropy
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
batch_size = 100
max_epochs = 100
early_stopping = EarlyStopping(patience=3)

model.fit(train_inputs,train_targets,batch_size=batch_size, epochs=max_epochs,
         validation_data=(validation_inputs,validation_targets),verbose=2,callbacks=[early_stopping])
```

```python
test_loss, test_accuracy = model.evaluate(test_inputs, test_targets)
```
